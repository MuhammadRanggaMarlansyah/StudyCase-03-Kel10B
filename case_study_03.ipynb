## 3. Outliers

```python
# garbungkan ketiga data
df = jakarta.merge(bogor, how="outer")
df = df.merge(bekasi, how="outer")

print(f'Jumlah baris sebelum memfilter outlier: {len(df)}')
# kolom = ["LT","LB", "KT", "KM","garasi","carport","sertifikat"]
for col in ['LT','LB', 'KT', 'KM','garasi','carport', 'harga']:
  Q1 = df[col].quantile(0.25)
  Q3 = df[col].quantile(0.75)
  IQR = Q3-Q1
  low_limit = Q1 - 1.5*IQR
  high_limit = Q3 + 1.5*IQR
  df = df[(df[col] >= low_limit) & (df[col] <= high_limit)]

print(f'Jumlah baris setelah memfilter outlier: {len(df)}')
```

## a. Kecendrungan Harga Rumah di Suatu Kota



```python
#Rata-rata harga rumah di ketiga kota
hr_jakarta = int(jakarta["harga"].mean())
print(f'Nilai rata-rata harga rumah di Jakarta: {hr_jakarta}')
hr_bogor = int(bogor["harga"].mean())
print(f'Nilai rata-rata harga rumah di bogor: {hr_bogor}')
hr_bekasi = int(bekasi["harga"].mean())
print(f'Nilai rata-rata harga rumah di bekasi: {hr_bekasi}')

h_ratarata = pd.Series([hr_jakarta, hr_bogor, hr_bekasi])
```

```python
if h_ratarata.max() == hr_jakarta:
  print("Kota Jakarta menjadi kota dengan nilai harga rata-rata rumah paling mahal")
elif h_ratarata.max() == hr_bogor:
  print("Kota bogor menjadi kota dengan nilai harga rata-rata rumah paling mahal")
elif h_ratarata.max() == hr_bekasi:
  print("Kota bekasi menjadi kota dengan nilai harga rata-rata rumah paling mahal")
```
